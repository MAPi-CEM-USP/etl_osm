```python
df23 = gpd.GeoDataFrame(
    df23,
    geometry=gpd.points_from_xy(df23["lon"], df23["lat"]),
    crs="EPSG:4326"
)
```

```python
from geopy.distance import geodesic

# Merge df_ano_novoand df23 on cod_escola
comparison = df23.merge(
    df_ano_novo,
    on='cod_escola',
    how='left',
    suffixes=('_23', '_ano_novo')
)

# Calculate distance between the two coordinate pairs (in meters)
# geodesic returns distance in km, multiply by 1000 for meters
comparison['distance_m'] = comparison.apply(
    lambda row: float('inf') if pd.isna(row['lat_ano_novo']) or pd.isna(row['lon_ano_novo']) else geodesic(
        (row['lat_ano_novo'], row['lon_ano_novo']),
        (row['lat_23'], row['lon_23'])
    ).meters,
    axis=1
)

# Add boolean column for proximity (within 100 meters)
comparison['coords_close'] = comparison['distance_m'] <= 100

# Display results
print(f"Total matches: {len(comparison)}")
print(f"Close matches (≤100m): {comparison['coords_close'].sum()}")
print(f"Different locations (>100m): {(~comparison['coords_close']).sum()}")
print(f"Not Matched: {(comparison['distance_m'] == float('inf')).sum()}")

comparison
```

```python
import folium
from folium import plugins

# Get bounds for the map
cidade = comparison#.loc[(comparison["nm_mun_ano_novo"] == "Belém")&(comparison["sg_uf_ano_novo"] == "PA")].copy()

# Convert lat/lon to numeric
cidade['lat_ano_novo'] = pd.to_numeric(cidade['lat_ano_novo'], errors='coerce')
cidade['lon_ano_novo'] = pd.to_numeric(cidade['lon_ano_novo'], errors='coerce')

# Remove rows with NaN coordinates
cidade = cidade.dropna(subset=['lat_ano_novo', 'lon_ano_novo'])

# Calculate center
center_lat = cidade['lat_ano_novo'].mean()
center_lon = cidade['lon_ano_novo'].mean()

# Create map with CartoDB basemap
m = folium.Map(
    location=[center_lat, center_lon],
    zoom_start=13,
    tiles='CartoDB positron'
)

# Plot close coordinates in green
for idx, row in cidade[cidade['coords_close']].iterrows():
    folium.CircleMarker(
        location=[row['lat_23'], row['lon_23']],
        radius=6,
        color='green',
        fill=True,
        fillColor='green',
        fillOpacity=0.7,
        popup=f"{row['nm_escola_23']}<br>Distance: {row['distance_m']:.1f}m<br>cod_escola: {row['cod_escola']}"
    ).add_to(m)

# Plot far coordinates in red
for idx, row in cidade[(~cidade['coords_close'])&(cidade['distance_m']!=float('inf'))].iterrows():
    folium.CircleMarker(
        location=[row['lat_23'], row['lon_23']],
        radius=6,
        color='red',
        fill=True,
        fillColor='red',
        fillOpacity=0.7,
        popup=f"{row['nm_escola_23']}<br>Distance: {row['distance_m']:.1f}m<br>cod_escola: {row['cod_escola']}"
    ).add_to(m)

# Plot far coordinates in red
for idx, row in cidade[(~cidade['coords_close'])&(cidade['distance_m']==float('inf'))].iterrows():
    folium.CircleMarker(
        location=[row['lat_23'], row['lon_23']],
        radius=6,
        color='black',
        fill=True,
        fillColor='black',
        fillOpacity=0.7,
        popup=f"{row['nm_escola_23']}<br>Distance: {row['distance_m']:.1f}m<br>cod_escola: {row['cod_escola']}"
    ).add_to(m)

# Add legend
legend_html = '''
     <div style="position: fixed; 
                     bottom: 50px; left: 50px; width: 200px; height: 100px; 
                     background-color: white; border:2px solid grey; z-index:9999; 
                     font-size:14px; padding: 10px">
     <p><b>Belém: Coordinate Comparison 2025 vs 2023</b></p>
     <p><i class="fa fa-circle" style="color:green"></i> Close (≤100m)</p>
     <p><i class="fa fa-circle" style="color:red"></i> Far (>100m)</p>
     </div>
'''
m.get_root().html.add_child(folium.Element(legend_html))

m
```
